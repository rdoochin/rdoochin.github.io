```python
%matplotlib inline
import pandas as pd

# housing_sp represents housing sale prices by city. (March 2008 - March 2020)
housing_sp = pd.read_csv('../data/archive/Sale_Prices_City.csv')

# msp represents the median sale price for houses sold in the US. (July 2005 - April 2020)
msp = pd.read_csv('../data/MSPUS.csv')
```

```python
### TIDY DATA FOR HOUSING_SP ###

# Melting
id_vars = ["RegionID",
           "RegionName",
           "StateName",
           "SizeRank"]

housing_sp = pd.melt(frame=housing_sp, id_vars=id_vars, var_name="Date", value_name="Price")
housing_sp = housing_sp.dropna()
housing_sp = housing_sp.sort_values(ascending=True, by=["RegionID", "Date"])
housing_sp
```
